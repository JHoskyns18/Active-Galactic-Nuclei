```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('Skyserver_CrossID4_23_2025 1_42_22 PM.csv', delimiter=',',skiprows=2)

OIII = data[:, 0]
HB = data[:, 1]
Ha = data[:, 2]
NII = data[:, 3]
```

```python
y = OIII/HB
x = NII/Ha
```

```python
plt.figure(figsize=(16, 10))
plt.loglog(x, y, label='',color='black',ls = '',marker='o')
x1 = np.linspace(0.005,1,1000)
y1 = np.exp(((0.61/(np.log(x1)-0.05)) + 1.3))
x2 = np.linspace(0.005,1.5,1000)
y2 = np.exp(((0.61/(np.log(x2)-0.47)) + 1.19))
x3 = np.linspace(0.83,5,1000)
y3 = np.exp((1.05*np.log(x3))+0.45)


plt.plot(x1,y1,'--',label='Theoretical curve (Kewley et al. 2001)')
plt.plot(x2,y2,label='Emperical curve (Kauffmann et al. 2003)')
plt.plot(x3,y3,color='red',label='Line to seperate LINER and Seyfert galaxies (Schawinski et al. 2007)')

plt.title('BPT Diagram')
plt.xlabel('Log NII/Ha')
plt.ylabel('Log OIII/HB')
plt.xlim(0.04,3)
plt.ylim(0.04,100)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
data2 = np.loadtxt('Skyserver_SQL4_24_2025 9_50_19 AM.csv', delimiter=',',skiprows=2)
```

```python
W1 = data2[:, 0]
W2 = data2[:, 1]
W3 = data2[:, 2]
```

```python
Y = W1 - W2
X = W2 - W3
Y1 = np.linspace(0.8,0.8,1000)
X1 = np.linspace(0,5,1000)
Y2 = np.linspace(-0.2,0.8,1000)
X2 = np.linspace(1.9,1.9,1000)
X3 = np.linspace(3.45,3.45,1000)
```

```python
plt.figure(figsize=(16, 10))
plt.plot(X,Y,label='',color='black',ls = '',marker='x')
plt.plot(X1,Y1,'--')
plt.plot(X2,Y2,'--')
plt.plot(X3,Y2,'--')
plt.title('WISE color-color diagram')
plt.xlabel('W2-W3 [mag]')
plt.ylabel('W1-W2 [mag]')
plt.grid(True)
plt.tight_layout()
plt.show()
```
